```python
def create_stamp(obj_grism_id):
    twod = fits.open((obj_grism_id+".2D.fits"), memmap=True)
    zfit = fits.open((obj_grism_id+ ".new_zfit.fits"), memmap=True)
    science = twod[5].data
    contam = twod[8].data
    wavelength = twod[9].data
    centre_stamp = wavelength_window(obj_grism_id, wavelength)
    
    #have to think about lines on edge of grism
    edge_flag = 0
    left_window = centre_stamp-10
    right_window = centre_stamp + 10
    
    if left_window<0:
        left_window = 0
        edge_flag = 1
    
    if right_window> 312:
        right_window = 312
        edge_flag=1
        
    plt.figure(figsize=((right_window-left_window)/2, 6.5))
    plt.imshow(science[0::, left_window:right_window], interpolation='nearest')
    img_array = (science-(contam+ (zfit[2].data) + (zfit[1].data)))
    plt.imshow(img_array[0::, left_window:right_window], interpolation = 'nearest')
    plt.imsave((obj_grism_id + ".stamp.png"), img_array[0::, left_window:right_window] )
    return edge_flag
```
